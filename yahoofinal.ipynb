```python
df5=[]
for url1 in link_all:
    #print(url)
    r=requests.get(url1,headers=headers)
    soup=BeautifulSoup(r.text,'lxml')  
    #物件編號
    car_id = []
    idnumber = soup.find_all('li',class_='last mei-g')
    idnumber1 = BeautifulSoup(str(idnumber), 'html.parser')
    idnumber2 = idnumber1.find_all('div',class_='mei-u dcontent')
    for z in idnumber2 :
        z = z.text.strip()
        car_id = z
    #車廠
    brand = soup.find_all('div',class_='itemhd')
    brand1 = BeautifulSoup(str(brand), 'html.parser')
    brand2 = brand1.find_all('a')[2]
    brand3 = brand2.text.strip()  
    result = ''.join(re.findall(r'[A-Za-z]', brand3))  
    car_brand = result.upper()
    ##車款
    try:   
      model = soup.find_all('div',class_='itemhd')
      model1 = BeautifulSoup(str(model), 'html.parser')  
      model2 = model1.find_all('a')[3]
      model3 = model2.text.strip()  
      result = ''.join(re.findall(r'[A-Za-z0-9]', model3))  
      car_model = result.upper()
    except IndexError:
      pass
    #照片網址        
    car_photo = soup.find('div', class_="imgwrap").img.get('src')
    #車輛價格
    car_price = []
    details = soup.find_all('li',class_='mei-g')
    details1 = BeautifulSoup(str(details), 'html.parser')
    details2 = details1.find_all('em',class_='price')
    for z in details2 :
        z = z.text.strip().strip(',')
        car_price = re.sub('元','',z)
    #刊登標題
    car_title = []
    title = soup.find_all('div', class_='prightn mei-u')
    title1 = BeautifulSoup(str(title), 'html.parser')
    title2 = title1.find_all('div',class_='rthd')
    for z in title2 :
        z = z.text.strip()  
        car_title = z  
    #車商名
    car_dealer = []    
    seller = soup.find_all('div',class_='pleftn mei-u')
    seller1 = BeautifulSoup(str(seller), 'html.parser')
    seller2 = seller1.find_all('div',class_='mbd')
    for z in seller2 :
        car_dealer.append(z.text.strip().split('\n')[0])
    #車輛(商)所在地
    car_address1 = []
    address = soup.find_all('li',class_='mei-g')
    address1 = BeautifulSoup(str(address), 'html.parser')
    address2 = address1.find('em', itemprop="address")
    car_address1.append(address2.text.strip())
    car_address = str(car_address1)[2:5]
    #車商地址 
    dealer_address = []
    addr = soup.find_all('li',class_='mei-g')
    addr1 = BeautifulSoup(str(addr), 'html.parser')
    addr2 = address1.find('em', itemprop="address")
    dealer_address.append(addr2.text.strip())

    tmp = [car_id,car_brand,car_model,car_photo,car_price,car_title,car_dealer,car_address,dealer_address]
    df5.append(tmp)
    df_detail_basic=pd.DataFrame(columns=['id','car_brand','car_model','car_photo','car_price','car_title','dealer_name','car_address','dealer_address'])           
    df6 = pd.DataFrame(df5 , columns = df_detail_basic.columns)
```

```python
##抓產品屬性
attr_columns = ['車源', '出廠年份', '刊登者身份', '外觀車色', '排氣量', '變速系統', '車門數', '行駛里程(KM)' 
                     , '是否在店', '承諾售價包括', '引擎燃料', '配備'] 
df = pd.DataFrame(columns=attr_columns)

for url1 in link_all:
    #print(url)
    r=requests.get(url1,headers=headers)
    soup=BeautifulSoup(r.text,'lxml')   
    equipment = soup.find_all('ul', id="itemAttrs")
    equipment2 = BeautifulSoup(str(equipment), 'html.parser')
    equipment3 = equipment2.find_all('tbody')[:3]

    #將元素分開
    x = []
    tmp2 = []
    for row in equipment3:
        tmp = row.find_all('td')
        tmp2 = BeautifulSoup(str(tmp), 'html.parser').text.strip()
        tmp3 = tmp2.replace('[','').replace(']','').replace("'",'')          
        x.append(tmp3)

    #清理元素
    y = []
    for i in x:
        tmp = i.split(",")
        tmp2 = [item.replace('[','').replace(']','').replace("'",'').strip()  for item in tmp]
        y.extend(tmp2)

    #將元素放到dataframe

    dictionary = dict(zip(y[::2], y[1::2]))
    #刪除空的dictionary元素
    for key in list(dictionary.keys()):
        if not dictionary.get(key):
            del dictionary[key]
    df1 = pd.DataFrame([dictionary])
    
    df1['排氣量'].replace('1200cc 以下', 1200, inplace=True)
    df1['排氣量'].replace('1201cc~1800cc', 1500, inplace=True)
    df1['排氣量'].replace('1801cc~2400cc', 2200, inplace=True)
    df1['排氣量'].replace('2401cc~3000cc', 2600, inplace=True)
    df1['排氣量'].replace('3001cc~3600cc', 3300, inplace=True)
    df1['排氣量'].replace('3601cc 以上', 4800, inplace=True)
    df1['車門數'].replace('二門', 2, inplace=True)
    df1['車門數'].replace('三門', 3, inplace=True)
    df1['車門數'].replace('四門', 4, inplace=True)
    df1['車門數'].replace('五門', 5, inplace=True)
    df = df.append(df1, ignore_index=True)
#df.head()
```
