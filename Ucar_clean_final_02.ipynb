```python
from google.colab import files
import numpy as np
import pandas as pd
files.upload()
```

```python
pd.set_option('display.max_columns', None)
df_Ucar = pd.read_csv('Ucar_all.csv',index_col=0)
```

```python
df_Ucar.drop(df_Ucar.loc[df_Ucar['car_price']=='面議','car_price'].index,inplace=True)
df_Ucar.drop(df_Ucar.loc[df_Ucar['car_price']=='已收訂','car_price'].index,inplace=True)
```

```python
df_Ucar['car_price'] = df_Ucar['car_price'].str.replace(',', '').astype(int)
```

```python
df_Ucar['car_year'] = df_Ucar['car_year'].str.split('領牌', expand=True)
```

```python
df_Ucar['car_year'] = df_Ucar['car_year'].str.replace(',', '').astype(int)
```

```python
df_Ucar.drop(df_Ucar.loc[df_Ucar['car_mileage']=='銀','car_mileage'].index,inplace=True)
df_Ucar.drop(df_Ucar.loc[df_Ucar['car_mileage']=='白','car_mileage'].index,inplace=True)
df_Ucar.drop(df_Ucar.loc[df_Ucar['car_mileage']=='天藍','car_mileage'].index,inplace=True)
df_Ucar.drop(df_Ucar.loc[df_Ucar['car_mileage']=='鐵灰','car_mileage'].index,inplace=True)
df_Ucar.drop(df_Ucar.loc[df_Ucar['car_mileage']=='黑','car_mileage'].index,inplace=True)
df_Ucar.drop(df_Ucar.loc[df_Ucar['car_mileage']=='藍','car_mileage'].index,inplace=True)
df_Ucar.drop(df_Ucar.loc[df_Ucar['car_mileage']=='紅','car_mileage'].index,inplace=True)
```

```python
df_Ucar['car_mileage'] = df_Ucar['car_mileage'].str.replace(',', '').astype(int)
```

```python
df_Ucar['car_color'] = df_Ucar['car_color'].str.split('色', expand=True)
```

```python
df_Ucar['car_cylinderVolume'] = df_Ucar['car_cylinderVolume'].str.replace(',', '').astype(int)
```

```python
df_Ucar['car_fuel'] = df_Ucar['car_fuel'].str.replace('油電混合', '油電')
```

```python
df_Ucar['car_brand'] = df_Ucar['car_brand'].str.replace('LAND', 'LAND ROVER')
```

```python
df_Ucar['source'] = 'U-car'
```

```python
df_Ucar.rename(columns={'ABS防鎖死煞車系統':'ABS','真皮座椅':'leather_seat','衛星導航設備':'navi_sys','天窗':'sunroof','影音系統':'IVI_sys','HID氣體放電頭燈':'HID_head','恆溫空調':'auto_AC','Keyless感應門鎖':'keyless','安全氣囊':'SRS','停車雷達系統':'parking_radar','四輪傳動':'car_driveMode'},inplace=True)
```

```python
del df_Ucar['CD音響']
```

```python
del df_Ucar['鋁合金輪圈']
```

```python
df_Ucar['car_driveMode'] = df_Ucar['car_driveMode'].astype(str)
```

```python
df_Ucar['car_driveMode'].dtype
```

```python
df_Ucar['car_driveMode'] = df_Ucar['car_driveMode'].str.replace('1', '4').astype(int)
```

```python
df_Ucar['car_driveMode'] = df_Ucar['car_driveMode'].astype(str)
```

```python
df_Ucar['car_driveMode'] = df_Ucar['car_driveMode'].replace('0', '2').astype(int)
```

```python
df_Ucar.to_csv('Ucar_clean_final_02.csv', mode='w', encoding='utf-8-sig')
```
