```python
from google.colab import files
import numpy as np
import pandas as pd
files.upload()
```

```python
pd.set_option('display.max_columns', None)
df_Ucar = pd.read_csv('Ucar_all.csv',index_col=0)
```

```python
df_Ucar.drop(df_Ucar.loc[df_Ucar['car_price']=='','car_price'].index,inplace=True)
df_Ucar.drop(df_Ucar.loc[df_Ucar['car_price']=='','car_price'].index,inplace=True)
```

```python
df_Ucar['car_price'] = df_Ucar['car_price'].str.replace(',', '').astype(int)
```

```python
df_Ucar['car_year'] = df_Ucar['car_year'].str.split('', expand=True)
```

```python
df_Ucar['car_year'] = df_Ucar['car_year'].str.replace(',', '').astype(int)
```

```python
df_Ucar.drop(df_Ucar.loc[df_Ucar['car_mileage']=='','car_mileage'].index,inplace=True)
df_Ucar.drop(df_Ucar.loc[df_Ucar['car_mileage']=='','car_mileage'].index,inplace=True)
df_Ucar.drop(df_Ucar.loc[df_Ucar['car_mileage']=='','car_mileage'].index,inplace=True)
df_Ucar.drop(df_Ucar.loc[df_Ucar['car_mileage']=='','car_mileage'].index,inplace=True)
df_Ucar.drop(df_Ucar.loc[df_Ucar['car_mileage']=='','car_mileage'].index,inplace=True)
df_Ucar.drop(df_Ucar.loc[df_Ucar['car_mileage']=='','car_mileage'].index,inplace=True)
df_Ucar.drop(df_Ucar.loc[df_Ucar['car_mileage']=='','car_mileage'].index,inplace=True)
```

```python
df_Ucar['car_mileage'] = df_Ucar['car_mileage'].str.replace(',', '').astype(int)
```

```python
df_Ucar['car_color'] = df_Ucar['car_color'].str.split('', expand=True)
```

```python
df_Ucar['car_cylinderVolume'] = df_Ucar['car_cylinderVolume'].str.replace(',', '').astype(int)
```

```python
df_Ucar['car_fuel'] = df_Ucar['car_fuel'].str.replace('', '')
```

```python
df_Ucar['car_brand'] = df_Ucar['car_brand'].str.replace('LAND', 'LAND ROVER')
```

```python
df_Ucar['source'] = 'U-car'
```

```python
df_Ucar.rename(columns={'ABS':'ABS','':'leather_seat','':'navi_sys','':'sunroof','':'IVI_sys','HID':'HID_head','':'auto_AC','Keyless':'keyless','':'SRS','':'parking_radar','':'car_driveMode'},inplace=True)
```

```python
del df_Ucar['CD']
```

```python
del df_Ucar['']
```

```python
df_Ucar['car_driveMode'] = df_Ucar['car_driveMode'].astype(str)
```

```python
df_Ucar['car_driveMode'].dtype
```

```python
df_Ucar['car_driveMode'] = df_Ucar['car_driveMode'].str.replace('1', '4').astype(int)
```

```python
df_Ucar['car_driveMode'] = df_Ucar['car_driveMode'].astype(str)
```

```python
df_Ucar['car_driveMode'] = df_Ucar['car_driveMode'].replace('0', '2').astype(int)
```
