```python
df1 = pd.read_csv('C:/Users/Lily/Desktop/Jupyter Workplace/BDSE19_PROJECT/final/car_isave_all_final_v2.csv', index_col=0)
df2 = pd.read_csv('C:/Users/Lily/Desktop/Jupyter Workplace/BDSE19_PROJECT/final/ABC_CAR_final.csv', index_col=0)
```

```python
#find the data that have price
df1 = df1.loc[df1['car_price'].str.contains('萬') == True]

df1['source'] = 'ISAVE'
df1['car_brand'] = df1['car_brand'].str.upper()
df1['car_model'] = df1['car_model'].str.upper()
df1['car_year'] = df1['car_year'].str.replace('年','')
df1['car_price'] = df1['car_price'].str.replace('萬','')
df1['car_price'] = pd.to_numeric(df1['car_price'])*10000

car_mileage = []
for i in df1['car_mileage']:
    if i != 'KM':
        tmp = float(str(i).replace('KM',''))#*1000
        car_mileage.append(tmp)

    else:
        tmp = np.NaN
        car_mileage.append(tmp)
df1['car_mileage'] = car_mileage
      
columns=['car_brand', 'car_model', 'car_year', 'car_mileage', 'car_price', 'source']
df1 = df1[columns]
```

```python
df1.to_csv('car_isave_cleaned_20210718.csv', encoding='utf_8_sig')
```
