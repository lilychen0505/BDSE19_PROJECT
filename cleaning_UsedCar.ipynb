```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
df_usedcar = pd.read_csv('UsedCar_final.csv',index_col=0)
```

```python
df_usedcar.rename(columns={'車款':'car_brand', '車型':'car_model', '年份':'car_year', '里程':'car_mileage','車輛價格':'car_price', '縣市':'car_address', '顏色':'car_color','車門數':'car_door','座位數':'car_seat', '排氣量':'car_cylinderVolume','驅動方式':'car_driveMode','燃料':'car_fuel','變速系統':'car_gear','刊登標題':'car_title','照片網址':'car_photo','車商名':'dealer_name','車商地址':'dealer_address'},inplace=True)
```

```python
df_usedcar.drop(df_usedcar.loc[df_usedcar['car_price']=='面議','car_price'].index,inplace=True)
```

```python
df_usedcar.drop(df_usedcar.loc[df_usedcar['car_price']=='已收訂','car_price'].index,inplace=True)
```

```python
df_usedcar['car_price']= df_usedcar['car_price'].str.split('萬', expand=True)
```

```python
df_usedcar['car_price']= pd.to_numeric(df_usedcar['car_price'] ,downcast='integer')
```

```python
df_usedcar['car_price']= df_usedcar['car_price'].astype(int)
```

```python
df_usedcar['car_price']= df_usedcar['car_price']*10000
```

```python
df_usedcar['car_price'].dtype
```

```python
df_usedcar['car_mileage']= df_usedcar['car_mileage'].str.split('公里', expand=True)
```

```python
df_usedcar.drop(df_usedcar.loc[df_usedcar['car_mileage'] == '0 ','car_mileage'].index,inplace=True)
```

```python
df_usedcar.drop(df_usedcar.loc[df_usedcar['car_mileage'] == '未知','car_mileage'].index,inplace=True)
```

```python
df_usedcar.reset_index(drop=True,inplace=True)
```

```python
df_usedcar['car_door']= df_usedcar['car_door'].str.split('門', expand=True)
```

```python
df_usedcar['car_seat']= df_usedcar['car_seat'].str.split('人座', expand=True)
```

```python
df_usedcar['car_driveMode']= df_usedcar['car_driveMode'].replace({'前輪驅動':'2WD','後輪驅動':'2WD','四輪驅動':'4WD'})
```

```python
df_usedcar['car_cylinderVolume']= df_usedcar['car_cylinderVolume'].str.split('cc', expand=True)
```

```python
df_usedcar['car_cylinderVolume']= pd.to_numeric(df_usedcar['car_cylinderVolume'] ,downcast='integer')
```

```python
df_usedcar['car_cylinderVolume'].dtype
```

```python
df_usedcar['car_brand'] = df_usedcar['car_brand'].str.upper()
```

```python
df_usedcar['car_model'] = df_usedcar['car_model'].str.upper()
```

```python
df_usedcar
```

```python
df_usedcar.to_csv('usedcar_clear.csv', mode='w',encoding='utf-8-sig')
```
