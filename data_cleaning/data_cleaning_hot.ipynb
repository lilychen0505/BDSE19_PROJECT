```python
df_hot = df_hot.reindex(['id','car_brand','car_model','car_year','car_mileage','car_price','car_color','car_cylinderVolume','car_driveMode',
                         'car_fuel','car_gear','verified','car_address','car_title','car_photo','dealer_name','dealer_address','sunroof',
                         'TPMS','SRS','keyless','ESP','ABS','cruise_sys','HID_head','isofix','electric_seat','auto_AC','navi_sys','IVI_sys','leather_seat',
                         'parking_radar','parking_camera','BSM','LDW'], axis=1)


```

```python
df_hot.loc[(df_hot['VCD'] | df_hot['DVD'] | df_hot['液晶螢幕']) == 1, 'IVI_sys'] = 1

df_hot['IVI_sys'] = df_hot['IVI_sys'].fillna(0).astype(int)
```

```python
df_hot = df_hot.drop(['SEQ','KM1','SALAMT1','SPLITSTR','BCOLOR','BODYTYPE','BRAND','CD','VCD','DVD','液晶螢幕','car_photo','BRANDCNM',
             'SEQNO','CERPHOTO','WARPHOTO','HOTPHOTO','PHOTOSIZE','BKCOLORCD','DEPOSIT','ASYSDT','GOLDMEMTYPE','電動後視鏡','電動尾門','電動滑門','全景天窗','方向盤換檔撥片',
             'CNUM','CERSTS','WARDUR','HOT_RECOMMEND','OUTFLG','SPRICE','CARYM','BODYTYPENM','TEL','PHOTOSTR','WARCD','KMCERT','NOTTAXI','CARCD','GOLDCODE','HOTCERCD','霧燈','鋁圈'], axis=1)
```

```python
df_hot['car_driveMode'].unique()
```

```python
df_hot['car_driveMode'].fillna(value=2, inplace=True)
```

```python
df_hot.replace({'car_driveMode': '二輪傳動'}, {'car_driveMode': 2}, inplace=True)

df_hot.replace({'car_driveMode': '四輪傳動'}, {'car_driveMode': 4}, inplace=True)
```

```python
df_hot.replace({'car_color': r'系$'}, {'car_color': ''}, regex=True, inplace=True)

df_hot.replace({'car_color': r'色$'}, {'car_color': ''}, regex=True, inplace=True)

df_hot.replace({'car_color': r'其他$'}, {'car_color': np.NaN}, regex=True, inplace=True)
```

```python
df_hot['car_cylinderVolume'] = df_hot['car_cylinderVolume'].astype(float)

df_hot['car_cylinderVolume']=df_hot['car_cylinderVolume']*1000

df_hot['car_cylinderVolume']=df_hot['car_cylinderVolume'].astype(int)
```

```python
df_hot['verified'] = df_hot['verified'].str.replace('Y','1')

df_hot['verified'] = df_hot['verified'].str.replace('N','0')

df_hot['source']='HOT'

df_hot['verified']=df_hot['verified'].astype(int)
```

```python
df_hot.to_csv('../csv/raw_csv/HOT_0728.csv', encoding='utf-8-sig')
```
