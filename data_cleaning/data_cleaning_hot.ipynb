```python
# 6267
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 80)
```

```python
df_hot1 = pd.read_csv('../csv/raw_csv/hot_car_0706.csv', index_col=0)
df_hot1
```

```python
df_hot2 = pd.read_csv('../csv/raw_csv/hot_car_0723.csv', index_col=0)
df_hot2 = df_hot2[['TSEQNO','car_fuel']]
```

```python
df_hot = df_hot1.merge(df_hot2,how='left', on='TSEQNO')
```

```python
df_hot
```

```python
df_hot.isnull().sum()
```

```python
df_hot = df_hot.drop(['SEQ','TSEQNO','KM1','SALAMT1','SPLITSTR','BCOLOR','BODYTYPE','BRAND',
             'SEQNO','CERPHOTO','WARPHOTO','HOTPHOTO','PHOTOSIZE','BKCOLORCD','DEPOSIT',
             'CNUM','CERSTS','WARDUR','HOT_RECOMMEND','OUTFLG','SPRICE'], axis=1)
```

```python
df_hot['car_driveMode'].unique()
```

```python
df_hot['car_driveMode'].fillna(value=2, inplace=True)
```

```python
df_hot.replace({'car_driveMode': '二輪傳動'}, {'car_driveMode': 2}, inplace=True)
```

```python
df_hot.replace({'car_driveMode': '四輪傳動'}, {'car_driveMode': 4}, inplace=True)
```

```python
df_hot['CCORLORNM'].unique()
```

```python
df_hot.replace({'CCORLORNM': r'系$'}, {'CCORLORNM': ''}, regex=True, inplace=True)
```

```python
df_hot.replace({'CCORLORNM': r'色$'}, {'CCORLORNM': ''}, regex=True, inplace=True)
```

```python
df_hot.replace({'CCORLORNM': r'其他$'}, {'CCORLORNM': np.NaN}, regex=True, inplace=True)
```

```python
df_hot['CCNUM_R'].unique()
```

```python
df_hot['CCNUM_R'] = df_hot['CCNUM_R'].astype(float)
```

```python
df_hot['CCNUM_R']=df_hot['CCNUM_R']*1000
```

```python
df_hot['CCNUM_R']=df_hot['CCNUM_R'].astype(int)
```

```python
df_hot['GEARTYPE'].unique()
```

```python
gear_mapping={'手自排':1,'手排':2,'自排':3}
df_hot['GEARTYPE'] = df_hot['GEARTYPE'].map(gear_mapping)
```

```python
df_hot['BODYTYPENM'].unique()
```

```python
df_hot['CARTYPENM'].unique()
```

```python
df_hot.reindex(columns=[])
```

```python
df_hot['CERCD'].unique()
```

```python
df_hot['CERCD'] = df_hot['CERCD'].str.replace('Y','1')
```

```python
df_hot['CERCD'] = df_hot['CERCD'].str.replace('N','0')
```

```python
df_hot['CERCD']=df_hot['CERCD'].astype(int)
```

```python
df_hot['source']='HOT'
```

```python
df_xxx = df_hot[['BRANDNM','CARTYPENM','CARYY','KM','SALAMT','CCORLORNM','CCNUM_R','car_driveMode','car_fuel','GEARTYPE','CERCD','source']]
```

```python
df_xxx.info()
```

```python
df_xxx.to_csv('../csv/cleaned_csv/HOT_clean_forPrice_0723.csv', encoding='utf-8-sig')
```
