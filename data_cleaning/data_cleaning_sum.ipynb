```python
df_sum = pd.read_csv('../csv/Sum_car_0707.csv', index_col=0)
df_sum
```

```python
df_sum.isnull().sum()
```

```python
df_sum.loc[df_sum['car_price'] == '已收訂', 'car_price'].count()
```

```python
df_sum = df_sum.drop(df_sum.loc[df_sum['car_price']=='已收訂','car_price'].index)
```

```python
df_sum['car_driveMode'].unique()
```

```python
df_sum = df_sum.replace({'car_driveMode': r'WD$'}, {'car_driveMode': ''}, regex=True)
```

```python
df_sum['car_door'].unique()
```

```python
df_sum = df_sum.replace({'car_door': r'門$'}, {'car_door': ''}, regex=True)
```

```python
df_sum['car_seat'].unique()
```

```python
df_sum = df_sum.replace({'car_price': r'萬$'}, {'car_price': ''}, regex=True)
```

```python
df_sum['car_color'].unique()
```

```python
df_sum = df_sum.replace({'car_color': r'系$'}, {'car_color': ''}, regex=True)
```

```python
df_sum['car_fuel'].unique()
```

```python
df_sum = df_sum.replace({'car_fuel': r'車$'}, {'car_fuel': ''}, regex=True)
```

```python
fuel_mapping = {'汽油':1,'柴油':2,'油電':3,'電動':4}
df_sum['car_fuel'] = df_sum['car_fuel'].map(fuel_mapping)
```

```python
df_sum['car_price'] = df_sum['car_price'].astype(float)
```

```python
df_sum['car_price'] = df_sum['car_price']*10000
```

```python
df_sum = df_sum.dropna(subset=['car_mileage'])
```
