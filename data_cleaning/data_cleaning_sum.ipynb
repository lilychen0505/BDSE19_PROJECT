```python
colsreindex=['id','car_brand','car_model','car_year','car_mileage','car_price','car_color','car_door','car_seat','car_cylinderVolumn','car_driveMode','car_fuel','car_gear','verified','car_address','car_title','dealer_name','dealer_address'
            ,'sunroof','TPMS','SRS','keyless','theft_alarm','central_lock','ESP','TCS','ABS','cruise_sys','HID_head','LED_head','isofix','electric_seat','auto_AC','navi_sys','IVI_sys','SW_button','leather_seat','parking_radar','parking_camera','BSM','ACC','LDW','source']
print(len(colsreindex))
```

```python
df_sum = df_sum.reindex(columns=colsreindex)
```

```python
df_sum
```

```python
df_sum['IVI_sys'] = df_sum['IVI_sys'].astype(int)
```

```python
df_sum['verified'] = df_sum['verified'].astype(int)
```

```python
df_sum.loc[(df_sum['VCD'] | df_sum['DVD'] | df_sum['中控螢幕'] | df_sum['數位電視'] | df_sum['液晶螢幕']) == 1, 'IVI_sys'] = 1
df_sum['IVI_sys'] = df_sum['IVI_sys'].fillna(0)
```

```python
df_sum = df_sum.drop(df_sum.loc[df_sum['car_price']=='已收訂','car_price'].index)

df_sum = df_sum.replace({'car_driveMode': r'WD$'}, {'car_driveMode': ''}, regex=True)

df_sum['car_driveMode']=df_sum['car_driveMode'].astype(int)

df_sum = df_sum.replace({'car_door': r'門$'}, {'car_door': ''}, regex=True)

df_sum.drop(df_sum.loc[df_sum['car_seat'] > 9,'car_model'].index, axis=0, inplace=True)

df_sum = df_sum.replace({'car_price': r'萬$'}, {'car_price': ''}, regex=True)

df_sum = df_sum.replace({'car_color': r'系$'}, {'car_color': ''}, regex=True)

df_sum = df_sum.replace({'car_color': r'色$'}, {'car_color': ''}, regex=True)
df_sum = df_sum.replace({'car_color': r'-$'}, {'car_color': ''}, regex=True)

df_sum = df_sum.replace({'car_fuel': r'車$'}, {'car_fuel': ''}, regex=True)

df_sum['car_price'] = df_sum['car_price'].astype(float)

df_sum['car_price'] = df_sum['car_price']*10000

df_sum['car_price'] = df_sum['car_price'].astype(int)
df_sum['car_gear'] = df_sum['car_gearType']
del df_sum['car_gearType']
df_sum = df_sum.dropna(subset=['car_mileage'])

mask = (df_sum['里程保證'] == 1) | (df_sum['三大保證'] == 1) | (df_sum['四大保固'] == 1) | (df_sum['非計程車']  == 1)

df_sum.loc[mask, 'verified' ] = 1
df_sum['car_mileage'] = df_sum['car_mileage'].astype(int)
df_sum['car_door'] = df_sum['car_door'].astype(int)
df_sum['source'] = 'SUM'
df_sum.drop(columns = ['其\u2003\u2003他', '鋁圈','全景天窗','電動尾門','電動滑門' ,'防眩後視鏡','外觀配備','電動後視鏡',
                       '後視鏡方向指示燈','車頂行李架','空力套件','燈光配備','霧燈','前霧燈',
                       'LED日行燈','LED尾燈','主動轉向頭燈','光感應式頭燈','頭燈水平調整','迎賓踏板燈','內裝配備',
                      '賽車椅','第三排座椅','電動窗','電動腰靠','後座分離傾倒','多功能行車資訊顯示器','液晶螢幕',
                      '抬頭顯示器','方向盤換檔控制','換檔撥片','左右獨立空調','後座出風口','車內粉塵過濾','後遮陽簾',
                      '4WD','備胎','一手車','渦輪增壓','晶片系統','核桃木飾板','氣壓懸吊','駕駛輔助','Push Start/按鈕式發車','感應式雨刷',
                      '車前距雷達','PVM/環景系統','APA/自動停車系統','StopStart/怠速熄火系統','HSA/斜坡起步輔助',
                      '安全輔助','EBD/煞車力道分配系統','BAS/煞車力道輔助系統','AEB/自動緊急煞車系統','兒童安全鎖','影音配備',
                      'CD','CD筒','VCD','DVD','USB插槽','中控螢幕','數位電視','藍牙通訊','音響主機','外部音源接入',
                       '三大保證','非計程車','四大保固','里程保證','YES認證'], axis=1, inplace=True)
```
