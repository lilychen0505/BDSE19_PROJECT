```python
# 6267
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 80)
```

```python
df_hot = pd.read_csv('../csv/hot_car_0706.csv', index_col=0)
df_hot
```

```python
df_hot.isnull().sum()
```

```python
df_hot = df_hot.drop(['SEQ','TSEQNO','KM1','SALAMT1','SPLITSTR','BCOLOR','BODYTYPE','BRAND',
             'SEQNO','CERPHOTO','WARPHOTO','HOTPHOTO','PHOTOSIZE','BKCOLORCD','DEPOSIT',
             'CNUM','CERSTS','WARDUR','HOT_RECOMMEND','OUTFLG','SPRICE'], axis=1)
```

```python
df_hot = df_hot.replace({'car_driveMode': '二輪傳動'}, {'car_driveMode': '2'}, regex=True)
```

```python
df_hot = df_hot.replace({'car_driveMode': '四輪傳動'}, {'car_driveMode': '4'}, regex=True)
```

```python
df_hot = df_hot.replace({'CCORLORNM': r'系$'}, {'CCORLORNM': ''}, regex=True)
```

```python
df_hot = df_hot.replace({'CCORLORNM': r'色$'}, {'CCORLORNM': ''}, regex=True)
```

```python
df_hot['CCNUM_R'] = df_hot['CCNUM_R'].astype(float)
```

```python
df_hot['CCNUM_R']=df_hot['CCNUM_R']*1000
```
