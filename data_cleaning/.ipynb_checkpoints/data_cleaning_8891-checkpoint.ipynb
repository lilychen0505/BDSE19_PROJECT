```python
df_8891_d.auto_make_date.str.split('</b>').apply(pd.Series)
```

```python
df_8891_d['car_year'].dtype
```

```python
df_8891_d['car_year'] = df_8891_d['car_year'].astype(int)
```

```python
df_8891_d['car_price'].dtype
```

```python
df_8891_d['car_mileage'] = df_8891_d['car_mileage'].astype(float)
```

```python
df_8891_d['auto_mileage_num'].str.contains('萬').dtype
```

```python
df_8891_d['car_mileage'] = df_8891_d['car_mileage'].astype(int)
```
