```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame(columns=['car_brand','car_model','car_year','car_mileage','car_price','source'])
```

```python
df_8891 = pd.read_csv('../csv/cleaned_csv/8891_clean_forPrice_0719.csv', header=0, index_col=0, names=df1.columns)
```

```python
df_HOT = pd.read_csv('../csv/cleaned_csv/HOT_clean_forPrice_0719.csv', header=0, index_col=0, names=df1.columns)
```

```python
df_SUM = pd.read_csv('../csv/cleaned_csv/SUM_clean_forPrice_0719.csv', header=0, index_col=0, names=df1.columns)
```

```python
df_isave = pd.read_csv('../csv/cleaned_csv/car_isave_cleaned_20210718.csv', header=0, index_col=0, names=df1.columns)
```

```python
df_abc = pd.read_csv('../csv/cleaned_csv/abc_car_cleaned_20210718_new2.csv', header=0, index_col=0, names=df1.columns)
```

```python
df_8891
```

```python
df_8891['car_model']= df_8891['car_model'].str.upper()
```

```python
df_isave.info()
```

```python
df_isave.dropna(axis=0, inplace=True)
```

```python
df_isave['car_year'] = df_isave['car_year'].astype(int)
df_isave['car_mileage'] = df_isave['car_mileage'].astype(int)
df_isave['car_price'] = df_isave['car_mileage'].astype(int)
```

```python
df_abc
```

```python
df_abc.info()
```

```python
df_abc = df_abc.replace({'car_mileage':','}, {'car_mileage':''}, regex=True)
```

```python
df_abc['car_mileage'] = df_abc['car_mileage'].astype('int64')
df_abc['car_price'] = df_abc['car_price'].astype('int64')
```

```python
df_final = pd.concat([df_HOT, df_SUM, df_8891, df_isave, df_abc])
```

```python
df_final = df_final.reset_index(drop=True)
```
