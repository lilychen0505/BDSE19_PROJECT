```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('../csv/raw_csv/car_isave_all_final_v2.csv', index_col=0)
```

### clean isave

```python
df1
```

```python
#find the data that have price
df1 = df1.loc[df1['car_price'].str.contains('萬') == True]

df1['source'] = 'ISAVE'
df1['car_brand'] = df1['car_brand'].str.upper()
df1['car_model'] = df1['car_model'].str.upper()
df1['car_year'] = df1['car_year'].str.replace('年','')
df1['car_year'] = pd.to_numeric(df1['car_year'])
df1['car_price'] = df1['car_price'].str.replace('萬','').astype(float)*10000
df1['car_cylinderVolume'] = df1['car_cylinderVolume'].str.replace('CC','')
df1['car_cylinderVolume'] = pd.to_numeric(df1['car_cylinderVolume'])
# df1['car_price'] = pd.to_numeric(df1['car_price'])*10000
df1 = df1.replace({'car_color': r'系$'}, {'car_color': ''}, regex=True)
df1 = df1.replace({'car_color': r'色$'}, {'car_color': ''}, regex=True)


car_mileage = []
for i in df1['car_mileage']:
    if i != 'KM':
        tmp = float(str(i).replace('KM',''))#*1000
        car_mileage.append(tmp)

    else:
        tmp = np.NaN
        car_mileage.append(tmp)
df1['car_mileage'] = car_mileage
df1.dropna(subset=['car_mileage','car_price'], inplace=True)  
columns=['car_brand', 'car_model', 'car_year', 'car_mileage', 'car_price', 'car_color', 'car_cylinderVolume', 'car_gear', 'verified', 'source',]
df1 = df1[columns]
```

```python
df1.head()
```

```python
df1.to_csv('../csv/cleaned_csv/car_isave_cleaned_20210722.csv', encoding='utf_8_sig')
```

### clean abc

```python
df2 = pd.read_csv('../csv/raw_csv/ABC_CAR_final.csv', index_col=0)
```

```python
pd.set_option('max_columns', None)
df2
```

```python
columns=['car_brand', 'car_model', 'car_year', 'car_mileage', 'car_price', 'car_color', 'car_seat','car_cylinderVolume', 'car_driveMode', 'car_fuel', 'car_gear', 'verified']
df2=df2[columns]
```

```python
df2['car_seat'].unique()
```

```python
#find the data that have price

df2 = df2.loc[df2['car_price'].apply(str).str.contains('Error') == False]
df2 = df2.loc[df2['car_price'].str.contains('萬') == True]

df2['source'] = 'ABC_CAR'
df2['car_mileage'] = df2['car_mileage'].str.replace('公里','')
df2 = df2.replace({'car_mileage':','}, {'car_mileage':''}, regex=True)
df2['car_mileage'] = df2['car_mileage'].astype(int)
df2 = df2.replace({'car_driveMode':'前輪驅動'}, {'car_driveMode':2}, regex=True)
df2 = df2.replace({'car_driveMode':'後輪驅動'}, {'car_driveMode':2}, regex=True)
df2 = df2.replace({'car_driveMode':'四輪驅動'}, {'car_driveMode':4}, regex=True)



df2 = df2.replace({'car_seat':'人$'}, {'car_seat':''}, regex=True)
df2.loc[df2['car_seat'].str.contains('以上') == True, 'car_seat'] = 8
df2['car_seat'] = df2['car_seat'].astype(float)
df2 = df2.replace({'car_color':'色$'}, {'car_color':''}, regex=True)

df2 = df2.replace({'car_cylinderVolume':r'L$'}, {'car_cylinderVolume': ''}, regex=True)
df2['car_cylinderVolume'] = ((df2['car_cylinderVolume'].astype(float))*1000).astype(int)


df2['car_brand'] = df2['car_brand'].str.upper()
df2['car_model'] = df2['car_model'].str.upper()
df2['car_year'] = df2['car_year'].str.split('年',1).str[0]
df2['car_year'] = pd.to_numeric(df2['car_year'])
df2['car_price'] = df2['car_price'].str.replace('萬','')
df2['car_price'] = pd.to_numeric(df2['car_price'])*10000
df2.loc[df2['verified'].str.contains('badge') == True, 'verified'] = 0
df2['verified'] = df2['verified'].astype(int)
df2
```

```python
df2.info()
```

```python
df2.to_csv('../csv/cleaned_csv/abc_car_cleaned_20210722.csv', encoding='utf_8_sig')
```
