```python
columns=['id', 'car_brand', 'car_model', 'car_year', 'car_mileage', 'car_price' , 'car_color',
         'car_door', 'car_seat', 'car_cylinderVolume', 'car_driveMode', 'car_fuel', 'car_gear',
         'verified', 'car_address', 'car_title', 'car_photo', 'dealer_name', 'seller_role', 
         'dealer_address', 'sunroof', 'TPMS', 'SRS', 'keyless', 'theft_alarm', 'central_lock',
         'ESP', 'TCS', 'ABS', 'cruise_sys', 'HID_head', 'LED_head', 'isofix', 'electric_seat',
         'auto_AC', 'navi_sys', 'IVI_sys', 'SW_button', 'leather_seat', 'parking_radar', 'parking_camera', 'BSM', 'ACC', 'LDW']

df = pd.DataFrame(columns=columns)

df1 = df1.loc[df1['car_price'].str.contains('萬') == True]

df1['source'] = 'ISAVE'
df1['car_brand'] = df1['car_brand'].str.upper()
df1['car_model'] = df1['car_model'].str.upper()
df1['car_year'] = df1['car_year'].str.replace('年','')
df1['car_price'] = df1['car_price'].str.replace('萬','')
df1['car_price'] = pd.to_numeric(df1['car_price'])*10000

car_mileage = []
for i in df1['car_mileage']:
    if i != 'KM':
        tmp = float(str(i).replace('KM',''))#*1000
        car_mileage.append(tmp)

    else:
        tmp = np.NaN
        car_mileage.append(tmp)
df1['car_mileage'] = car_mileage
      
df_isave = df.append(df1, ignore_index=True)
df_isave = df_isave[columns]
```

```python
#df_isave['car_year'] = df_isave['car_year'].astype('Int64')
df_isave['car_color'] = df_isave['car_color'].str.replace('色','').str.replace('系','')
df_isave['car_door'] = df_isave['car_door'].astype('Int64')
df_isave['car_seat'] = df_isave['car_seat'].astype('Int64')
df_isave['car_cylinderVolume'] = df_isave['car_cylinderVolume'].str.replace('CC','')
#df_isave['car_cylinderVolume'] = df_isave['car_cylinderVolume'].astype('Int64') 不能改先不改
df_isave['verified'] = df_isave['verified'].astype('Int64')
```

```python
columns=['id', 'car_brand', 'car_model', 'car_year', 'car_mileage', 'car_price' , 'car_color',
         'car_door', 'car_seat', 'car_cylinderVolume', 'car_driveMode', 'car_fuel', 'car_gear',
         'verified', 'car_address', 'car_title', 'car_photo', 'dealer_name', 'seller_role', 
         'dealer_address', 'sunroof', 'TPMS', 'SRS', 'keyless', 'theft_alarm', 'central_lock',
         'ESP', 'TCS', 'ABS', 'cruise_sys', 'HID_head', 'LED_head', 'isofix', 'electric_seat',
         'auto_AC', 'navi_sys', 'IVI_sys', 'SW_button', 'leather_seat', 'parking_radar', 'parking_camera', 'BSM', 'ACC', 'LDW']

df = pd.DataFrame(columns=columns)

df2 = df2.loc[df2['car_price'].apply(str).str.contains('Error') == False]
df2 = df2.loc[df2['car_price'].str.contains('萬') == True]

df2['source'] = 'ABC_CAR'
df2['car_mileage'] = df2['car_mileage'].str.replace('公里','')
df2['car_brand'] = df2['car_brand'].str.upper()
df2['car_model'] = df2['car_model'].str.upper()
df2['car_year'] = df2['car_year'].str.split('年',1).str[0]
df2['car_price'] = df2['car_price'].str.replace('萬','')
df2['car_price'] = pd.to_numeric(df2['car_price'])*10000

df_abc = df.append(df2, ignore_index=True)
df_abc = df_abc[columns]
```

```python
df_abc['id'] = df_abc['id'].str.replace('物件編號','')
df_abc['car_mileage'] = df_abc['car_mileage'].str.replace(',','')
df_abc['car_price'] = df_abc['car_price'].astype('Int64')
df_abc['car_color'] =df_abc['car_color'].str.replace('色','').str.replace('系','')
df_abc['car_seat'] =df_abc['car_seat'].str.replace('人','').str.replace('含以上','').str.replace('(','').str.replace(')','')
df_abc['car_cylinderVolume'] = df_abc['car_cylinderVolume'].str.replace('L','')
df_abc['car_cylinderVolume'] = pd.to_numeric(df_abc['car_cylinderVolume'])*1000
df_abc['car_driveMode']=df_abc['car_driveMode'].str.replace('後輪驅動','2').str.replace('前輪驅動','2').str.replace('四輪驅動','4')
```
